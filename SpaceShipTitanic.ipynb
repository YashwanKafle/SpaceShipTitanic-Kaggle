```python
'''
Missing Values for HomePlanet
'''
def missingValuesHomePlanet(df):
    temp = df['HomePlanet'].value_counts()
    p = temp.values
    p = p/(len(df)-df['HomePlanet'].isna().sum())
    size = df['HomePlanet'].isna().sum()
    df.loc[df['HomePlanet'].isna(),'HomePlanet'] = np.random.choice(['Europa', 'Earth', 'Mars'],size, p = p)
    return df
```

```python
'''
 Cryo Sleep can be False for Nan Values
'''

def CryoSleepNaN(df):
    df['CryoSleep'].fillna(False,inplace=True)
    return df
```

```python
'''
Cabin Deck/Num/Side
'''
def cabinFeature(df):
    df[['Deck', 'Cabin_num','Cabin_side']] = df['Cabin'].str.split('/',expand=True)
    return df
```

```python
def deckNaN(df):
    temp = df['Deck'].value_counts()
    p = temp.values
    p = p/(len(df)-df['Deck'].isna().sum())
    size = df['Deck'].isna().sum()
    df.loc[df['Deck'].isna(),'Deck'] = np.random.choice(['F','G','E','B','C','D','A','T'],size, p = p)
    return df

def cabinNumNaN(df):
    df['Cabin_num'].fillna(0,inplace=True)
    return df

def cabinSideNaN(df):
    df.loc[df['Cabin_side'].isna(), 'Cabin_side'] = np.random.choice(['P','S'],df['Cabin_side'].isna().sum(),
                                                                p = [0.5,0.5])
    return df
```

```python
df_train = deckNaN(df_train)
df_train = cabinNumNaN(df_train)
df_train = cabinSideNaN(df_train)
```

```python
df_train.head()
```

```python
'''
Missing values for Destination
'''
def missingValuesDestination(df):
    temp = df['Destination'].value_counts()
    p = temp.values
    p = p/(len(df)-df['Destination'].isna().sum())
    size = df['Destination'].isna().sum()
    df.loc[df['Destination'].isna(),'Destination'] = np.random.choice(['TRAPPIST-1e', '55 Cancri e', 'PSO J318.5-22'],size, p = p)
    return df
```

```python
def luxiries(df):
    for i in range(len(df)):
        if df['CryoSleep'][i] == True:
            df['RoomService'][i] = 0
            df['FoodCourt'][i] = 0
            df['ShoppingMall'][i] = 0
            df['Spa'][i] = 0
            df['VRDeck'][i] = 0
  

    df['RoomService'].fillna(df['RoomService'].mean(),inplace=True)
    df['FoodCourt'].fillna(df['FoodCourt'].mean(),inplace=True)
    df['ShoppingMall'].fillna(df['ShoppingMall'].mean(),inplace=True)
    df['Spa'].fillna(df['Spa'].mean(),inplace=True)
    df['VRDeck'].fillna(df['VRDeck'].mean(),inplace=True)
    df['Luxiries'] = df['RoomService'] + df['FoodCourt'] + df['ShoppingMall'] + df['Spa'] + df['VRDeck']
    return df
```
